```python
# Module setup

import numpy as np
import scipy as sp
import matplotlib.pyplot as py
from pylab import *
# For netCDF go to http://www.lfd.uci.edu/~gohlke/pythonlibs/ and dowlnoad the nedCDF library
from netCDF4 import Dataset, num2date, date2num
from IPython.html.widgets import *

```

```python
# Set parameters

C      =    1             # Conductance

E_L    =  -80             # Leakage equilibrium
E_Na   =   60             # Sodium equilibrium
E_K    =  -90             # Potassium equilibrium

g_L    =    8             # Leakage current
g_Na   =   20             # Sodium current
g_K    =   10             # Potassium current

V_1by2 =  -20             
V_nby2 =  -25
k      =   15              
tau    = 0.13             
k_n    =    5

```

```python
# Differential equations:

def V_dot(x, y, I):
    return ((1 / C)*(I-g_L*(x-E_L)))-g_Na*((x-E_Na)/( 1 + np.exp( (V_1by2-x)/k)))-g_K*y*(x-E_K)

def n_dot(x, y, I):
    return (1/tau)*((1/(1+np.exp((V_nby2-x)/k_n)))-y)
    
```

```python
# Euler method for two differential equations

def euler2d(x_0,y_0,t_max,step_size, I):
    # Defining variables and vectors
    time = np.arange(0,t_max+dt,dt)                         # vector containing all time steps
    x =         []                                          # empty vector for target function
    y =         []
    tNow =      0                                           # control variable for time
    dx =        0                                           # storing change in target function
    dy =        0
    x_last =    0                                           # storing last value of target function 
    y_last =    0
    x_new =     0                                           # storing new value of target function 
    y_new =     0

    for tNow in time:
        if tNow == 0:                                       # if simulation just started ...
            x.append(x_0)                                   # then target function starts at x_0
            y.append(y_0)
        else:                                               
            x_last = x[-1]
            y_last = y[-1]
            dx = V_dot(x_last, y_last, I) * dt              # (differential equation)
            dy = n_dot(x_last, y_last, I) * dt              # (differential equation)
            x_new = x_last + dx
            y_new = y_last + dy
            x.append(x_new)
            y.append(y_new)
    
    return x,y
        
```

Initial Terms:

V_dot $ =  \frac{1}{C}(I-g_L (x-E_L)))-g_{Na} \frac{(x-E_Na)}{( 1 + e^{( (V_1by2-x)/k)})}-g_K y (x-E_K) $

n_dot $ = \frac{1}{\tau} (\frac{1}{1+e^{V_nby2-x / k_n}}-y) $

Now we set the differential equation V_dot to 0 (as we seek the nullcline) and in the first term we set:

$ (I-g_L (x-E_L)))-g_{Na} \frac{(x-E_Na)}{( 1 + e^{( (V_1by2-x)/k)})} = TERM_1 $

$ 0 = \frac{1}{C}*TERM_1 -g_K y (x-E_K) $

$ y = \frac{TERM_1}{g_K (x-E_K)} $

Now we set the differential equation n_dot to 0 (as we seek the nullcline) and in the second term we set:

$ \frac{1}{1+e^{V_nby2-x / k_n}} = TERM_2 $

$ 0 = \frac{1}{\tau} (TERM_2-y) $

$ y = TERM_2 $

Now we can plot the nullclines!
 



```python
def V_nullcline(x,I):
    return ((I-g_L*(x-E_L))-g_Na*((x-E_Na)/( 1 + np.exp( (V_1by2-x)/k)))) / (g_K*(x-E_K))

def n_nullcline(x,I):
    return 1/(1+np.exp((V_nby2-x)/k_n))

    
```

```python
# Create a grid for the 2D system

xlength = np.linspace(-90,20,100) # given by task [x axis: -90mV to 20mV]
ylength = np.linspace(-0.4,1,100) # given by task [y axis: -0.4mV to 1mV]

X,Y = meshgrid(xlength,ylength)

# Create interactive plot

def Inap(I):
    figure(figsize=(17,5))
    
    # Differential equations
    U = V_dot(X,Y,I)
    V = n_dot(X,Y,I)
    
    # Use fancy streamplot function
    streamplot(X,Y,U,V,color='b', density=1)
    xlim(-90,20)
    xlabel('Voltage in mV', size=15)
    ylim(-0.4,1)
    ylabel('Gating Variable n', size=15)
    title('$ I_{Na,p} + I_k Model for I = %s $' % I, size=15)
    plot([-90,20],[0,0],'k-', linestyle = 'dashed')
    grid()
    
    V_null = V_nullcline(xlength,I)
    plot(xlength,V_null, 'g', linewidth=2, label='V-Nullcline')
    n_null = n_nullcline(xlength,I)
    plot(xlength,n_null, 'r', linewidth=2, label='n-Nullcline')
    legend()
    
# set interaction parameters
interact(Inap, I=(0,30,5))
    
```

```python
# Defining parameters for euler
t_max =    1000.0               # end of simulation
dt =       0.01                 # step size of approximation 


# function doing: euler method 
def trajectories(I,x,y, show_Nullclines):
    x_eul, y_eul = euler2d(x,y,t_max,dt,I)
    
    figure(figsize=(17,5))
    plot(x_eul,y_eul)
    xlim(-90,20)
    xlabel('Voltage in mV', size=15)
    ylim(-0.4,1)
    ylabel('Gating Variable n', size=15)
    title('$ I_{Na,p} + I_k Model$ $trajectories$ $for$ $I = %s $ $;$ $Starting$ $at$ $(%s,%s)$' % (I,x,y), size=15)
    plot(x,y,'ro',label='start point')
    plot([-90,20],[0,0],'k-', linestyle = 'dashed')   # line through y=0
    
    #plot nullclines again:
    if show_Nullclines:
        V_null = V_nullcline(xlength,I)
        plot(xlength,V_null, 'g')
        n_null = n_nullcline(xlength,I)
        plot(xlength,n_null, 'r')

# use interaction for trajectories
interact(trajectories, I=(0,30,5), x=(-90,20,1),y=(-0.4,1,0.1), show_Nullclines=True)   
```
